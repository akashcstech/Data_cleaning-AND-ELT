# DATA VIDUALISATION- EDA

```python
import matplotlib.pyplot as plt
```

# RATING 

```python
data['Rate out of 5'].value_counts().sort_index().plot(kind="bar", x='Rate out of 5', figsize=(12, 8))
plt.title('Rating popularity', fontsize= 14, fontweight='bold')
plt.xlabel('Rating',fontsize=12)
plt.ylabel('frequency', fontsize=12)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

# "ONLINE ORDER" AND "BOOKING TABLE" COMPARISION

```python
fig, (plot1, plot2) = plt.subplots(1, 2 , figsize=(12,5))
# Online Table
online_counts = data['Available_Online_order'].map({True: "Available", False: "Not Available"}).value_counts()
plot1.bar(online_counts.index, online_counts.values, color='skyblue')
plot1.set_title("Online Order", fontsize=14, fontweight="bold")
plot1.set_xlabel("Online Order Availability", fontsize=12)
plot1.set_ylabel("Count", fontsize=12)
plot1.grid(axis='y', alpha=0.2)

# Book Table
book_table = data['Available_Book_table'].map({True: "Available", False: "Not Available"}).value_counts()
plot2.bar(book_table.index, book_table.values)
plot2.set_title("Book Table", fontsize=14, fontweight="bold")
plot2.set_xlabel("Book Order Availability", fontsize=12)
plot2.set_ylabel("Count", fontsize=12)
plot2.grid(axis='y', alpha=0.2)
plt.tight_layout()
plt.show()
```

# "ONLINE ORDER" VS "BOOKING TABLE" USING HEATMAP

```python
import seaborn as sns

plt.figure(figsize=(12,8))
sns.heatmap(pd.crosstab(data['Available_Online_order'], data['Available_Book_table']), annot=True, cmap="Blues")
plt.title("Heatmap: Online Order vs Book Table", fontsize=14, fontweight="bold")
plt.xlabel("Avaliable_Book_table", fontsize=12)
plt.ylabel("Avaliable_Online_order", fontsize=12)
plt.show()
```

# RATING VS POPULARITY COMPARISION

```python
plt.figure(figsize=(12, 8))
plt.scatter(data['Rate out of 5'],data['Votes of popularity'], alpha=0.7, s=100, color='blue', edgecolors='black')
plt.xlabel('Rate out of 5')
plt.ylabel('Votes of popularity')
plt.title('Rating vs Votes Scatter Plot', fontsize=14, fontweight="bold")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

# RESTAURANT DISTIBUTION BY LOCATION ANALYSIS

```python
plt.figure(figsize=(12, 8))
data['Location'].value_counts().sort_values(ascending=False).head(15).sort_values(ascending=True).plot(kind='barh', color='skyblue', edgecolor='black')
plt.title('Restaurant Distribution by Location', fontsize=14, fontweight='bold')
plt.xlabel('Number of Restaurants')
plt.ylabel('Location')
plt.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.show()
```

# RESTAURANT TYPE DISTRUIBUTION ANALYSIS

```python
rest_types = data['Rest_type'].str.split(',').explode()
rest_types = rest_types.str.strip()
plt.figure(figsize=(12, 8))
rest_types.value_counts().plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Restaurant Type Distribution', fontsize=14, fontweight='bold')
plt.xlabel('Restaurant Type',  fontsize=12)
plt.ylabel('Count',  fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

# COST FOR TWO PEOPLE AND ONE PERSON - SIDE BY SIDE ANALYSIS

```python
fig, (plot1, plot2) = plt.subplots(1, 2, figsize=(16, 6))

# Cost for two people
plot1.hist(data['Approx_cost(for two people)'],bins=35, alpha=0.7, color='skyblue', edgecolor='black')
plot1.set_title('Cost for Two People', fontweight='bold')
plot1.set_xlabel('Cost', fontsize=12)
plot1.set_ylabel('Frequency', fontsize=12)
plot1.grid(True, alpha=0.3)

# Cost for one person
plot2.hist(data['Approx_cost(for one people)'],bins=35, alpha=0.7, color='pink', edgecolor='black')
plot2.set_title('Cost for One Person', fontweight='bold')
plot2.set_xlabel('Cost')
plot2.set_ylabel('Frequency')
plot2.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
